```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("Cleaned_Sales.csv")
```

# Identify the top 10 customers who generated the highest overall losses for the company.

```python
loss_customer = df.groupby(
    ["Customer_ID","Customer_Name"]
)["Profit"].sum().sort_values().head(10)

print(loss_customer)
```

# Visualizationn

```python
loss_customer.plot(kind="barh")

plt.title("Top 10 Loss Making Customers")

plt.savefig("Top 10 Loss_Making_Customer.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Date Formate Convert

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Ship_Date"] = pd.to_datetime(df["Ship_Date"])
```

# Shipping Delay Analysis [Find the average delivery time for each shipping mode.]

```python
df["Delivery_Days"] = (
    df["Ship_Date"] - df["Order_Date"]
).dt.days

shipping = df.groupby("Ship_Mode")["Delivery_Days"].mean()

print(shipping)
```

# Visualization

```python
shipping.plot(kind="bar")

plt.title("Average Delivery Time")

plt.savefig("Average_Delivery_Time.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Late Delivery Impact [Determine whether deliveries taking more than 3 days have a higher return rate.]

```python
df["Delivery_Status"] = np.where(
    df["Delivery_Days"]>3,
    "Late",
    "On_Time"
)

late = df.groupby(
"Delivery_Status"
).agg(
Orders=("Order_ID","count"),
Returns=("Returned",
lambda x:(x=="Yes").sum())
)

late["Return Rate"] = (
late["Returns"]/late["Orders"])*100

print(late)
```

# Visualization

```python
late["Return Rate"].plot(kind="bar")

plt.savefig("Late_Delivery_Impact.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Salesperson Performance [Highest Salesperson] [Lowest Profit Margin Salesperson]

```python
salesperson = df.groupby(
"Retail_Sales_People"
).agg(
Sales=("Sales","sum"),
Profit=("Profit","sum")
)

salesperson["Profit Margin"] = (
salesperson["Profit"]/
salesperson["Sales"]
)*100

print(salesperson)
```

# Visualization

```python
salesperson["Sales"].plot(kind="bar")

plt.savefig("Sales_Person_Parformance.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Check whether the top customers contribute nearly 80% of total sales.

```python
pareto = df.groupby(
"Customer_ID"
)["Sales"].sum().sort_values(
ascending=False
).reset_index()

pareto["Running_Total"] = pareto["Sales"].cumsum()

pareto["Percentage"] = (
pareto["Running_Total"]/
pareto["Sales"].sum()
)*100

pareto.head()
```

# Visualization

```python
plt.plot(
pareto["Percentage"]
)

plt.title(
"Pareto Analysis"
)

plt.savefig("customers contribute nearly 80%.png", dpi=300, bbox_inches="tight")

plt.show()
```
