```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("Cleaned_Sales.csv")
```

# Dataset Overview

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Total_Sales

```python
print("Total Sales :", round(df["Sales"].sum(),2))
```

# Total_Profit

```python
print("Total Profit :", round(df["Profit"].sum(),2))
```

# Total_Orders

```python
print("Total Orders :", df["Order_ID"].nunique())
```

# Total_Customer

```python
print("Total Customers :", df["Customer_ID"].nunique())
```

# Total_Products

```python
print("Total Products :", df["Product_ID"].nunique())
```

# Catogory_Wise Sales 

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print(category_sales)
```

# Visualization

```python
category_sales.plot(kind="bar")

plt.title("Category Wise Sales")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.savefig("Category_Wise_Sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Sub-Category-Wise Profit

```python
sub_profit = df.groupby("Sub-Category")["Profit"].sum().sort_values()

print(sub_profit)
```

# Visualization

```python
sub_profit.plot(kind="barh")

plt.title("Sub Category Wise Profit")

plt.savefig("Sub_Category_Wise_Sales.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Resion-Wise Sales

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

# Visualization

```python
region_sales.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")
plt.title("Region Wise Sales")

plt.savefig("Region_Wise_Sales.png", dpi=300, bbox_inches="tight")
plt.show()
```

# State-Wise Top Sales

```python
top_states = df.groupby("State")["Sales"].sum().nlargest(10)

print(top_states)
```

# Visualization

```python
top_states.plot(kind="bar")

plt.title("Top 10 States by Sales")

plt.savefig("State_Wise_Sales.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Segment-Wise Sales

```python
segment_sales = df.groupby("Segment")["Sales"].sum()

print(segment_sales)
```

# Visualization

```python
segment_sales.plot(kind="bar")

plt.title("Segment Wise Sales")

plt.savefig("Segment_Wise_Sales.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Shipping-Mode Analysis

```python
ship = df.groupby("Ship_Mode")["Sales"].sum()

print(ship)
```

# Visualization

```python
ship.plot(kind="bar")

plt.title("Sales by Ship Mode")

plt.savefig("Sales_By_Ship_Mode.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Monthly Sales Trend

```python
monthly = df.groupby("Month_Name")["Sales"].sum()

print(monthly)
```

# Visualization

```python
monthly.plot(marker="o")

plt.title("Monthly Sales Trend")

plt.xticks(rotation=45)

plt.savefig("Monthly_Sales_Trend.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Profit Distribution

```python
plt.hist(df["Profit"], bins=30)

plt.title("Profit Distribution")

plt.savefig("Profit_Distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Discount Vs Profit

```python
plt.scatter(df["Discount"], df["Profit"])

plt.title("Discount vs Profit")

plt.xlabel("Discount")

plt.ylabel("Profit")

plt.savefig("Discount Vs Profit.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Correlation

```python
corr = df[["Sales","Profit","Discount","Quantity"]].corr()

print(corr)
```

# Heatmap

```python
sns.heatmap(corr, annot=True)

plt.savefig("Correlation.png", dpi=300, bbox_inches="tight")
plt.show()
```
