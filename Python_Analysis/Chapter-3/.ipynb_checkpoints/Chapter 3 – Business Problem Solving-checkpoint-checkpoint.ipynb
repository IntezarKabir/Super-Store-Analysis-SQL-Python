```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("Cleaned_Sales.csv")
```

# Convert Datetime Type

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

# Calculate the year-over-year (YoY) growth in Sales and Profit.

```python
yoy = df.groupby(df["Order_Date"].dt.year).agg(
    Total_Sales=("Sales", "sum"),
    Total_Profit=("Profit", "sum")
)

print(yoy)
```

# visualization

```python
yoy.plot(figsize=(10,5), marker="o")

plt.title("Year-wise Sales and Profit")

plt.savefig("Year-Wise-Sales And Profit.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Seasonality (Weekend Effect) [Highest Sales Day, Highest Returned Orders Day]

```python
weekday=df.groupby("Day_Name").agg(
Sales=("Sales","sum"),
Returns=("Returned",lambda x:(x=="Yes").sum())
)

print(weekday)
```

# Visualization

```python
weekday["Sales"].plot(kind="bar")

plt.savefig("Day_Name_Highest_Return_Order.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Find the best-performing quarter.

```python
quarter=df.groupby(
["Year","Quarter"]
)["Sales"].sum().sort_values(
ascending=False
)

print(quarter)
```

# Visualization

```python
quarter.plot(kind="bar")
plt.savefig("Best_Performance_Quarter.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Discount Trap [Find categories where heavy discount caused losses.]

```python
loss=df.groupby(
["Category","Sub-Category"]
).agg(
Avg_Discount=("Discount","mean"),
Total_Profit=("Profit","sum")
)

loss=loss[
loss["Total_Profit"]<0
]

print(loss)
```

# Visualization

```python
loss["Total_Profit"].plot(
kind="barh"
)
plt.savefig("Discount_Trap.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Return Problem [Find Return Rate by Region]

```python
region=df.groupby("Region").agg(
Orders=("Order_ID","count"),
Returns=("Returned",
lambda x:(x=="Yes").sum())
)

region["Return Rate"]=(
region["Returns"]/
region["Orders"]
)*100

print(region)
```

# Visualization

```python
region["Return Rate"].plot(
kind="bar"
)
plt.savefig("Return_Rate.png", dpi=300, bbox_inches="tight")

plt.show()
```
